# a. Cloak Summary Data

<section id="sectioncloak"> </section>


```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis, iqr, scoreatpercentile

file_path = "Invisibility_Cloak.xlsx"
data_sheet = pd.read_excel(file_path, sheet_name="Invisibility Cloak")

list_num = data_sheet["Cloak"]

mean_value = np.mean(list_num)
median_value = np.median(list_num)
mode_value = list_num.mode().iloc[0] 
std = np.std(list_num)
variance = np.var(list_num)
skewness_value = skew(list_num)
kurtosis_value = kurtosis(list_num)

q1 = np.percentile(list_num, 25)
q2 = np.percentile(list_num, 50)
q3 = np.percentile(list_num, 75)

iqr_value = iqr(list_num)

d9 = scoreatpercentile(list_num, 90)
p95 = scoreatpercentile(list_num, 95)

std_error_skewness = np.sqrt(6 * len(list_num) * (len(list_num) - 1) / ((len(list_num) - 2) * (len(list_num) + 1) * (len(list_num) + 3)))

results_df_cloak = pd.DataFrame({
    "Cloak Summary Data": ["", "", "", "", "", "", "", "", "", "", "", "", "", ""],
    "Data": ["Valid", "Mode", "Mean", "Median", "Standard Deviation", "Variance", "Skewness", "Standard Error of Skewness", "Kurtosis", "Q1", "Q2", "Q3", "D9", "P95"],
    "Score": [len(list_num), mode_value, mean_value, median_value, std, variance, skewness_value, std_error_skewness, kurtosis_value, q1, q2, q3, d9, p95]
})

results_df_cloak["Score"] = results_df_cloak["Score"].apply(lambda x: "{:.3f}".format(x))
results_df_cloak
```
